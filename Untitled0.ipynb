# Introduction



The summer of 2023 saw a major wildfire on Rhodes Island (Greece), which severely affected Mediterranean shrub-and-forest ecosystems, altered surface conditions and increased the vulnerability of soils to erosion and runoff. Wildfires in such environments are of high ecological and societal concern because they not only destroy vegetation and habitat, but also perturb carbon and nutrient cycles, modify land-surface albedo and hydrology, and may trigger secondary hazards such as landslides or flash floods in the post-fire period. Monitoring and quantifying changes induced by wildfire events is therefore critical from both environmental and climate-resilience perspectives.

In this context, synthetic aperture radar (SAR) observations provide an especially valuable tool for investigation. The Sentinel1 mission, operated by European Space Agency (ESA), acquires C-band dual-polarisation backscatter imagery at frequent revisit intervals, offering day-and-night as well as all-weather capability. Unlike optical sensors, SAR is unaffected by smoke plumes or cloud cover, which often hamper wildfire monitoring during or immediately after the event. The temporal evolution of Sentinel-1 backscatter can therefore be used to detect surface change, vegetation removal, soil exposure and structural alteration of the land surface. For example, Sentinel-1 time series have been successfully applied to near-real-time wildfire progression monitoring using convolutional-neural-network frameworks (see Ban et al., 2020). Moreover, SAR change-detection methods have been exploited to assess fire-induced forest loss in tropical forest contexts (e.g., Chvez et al., 2023) and to map burn severity via changes in cross-polarised backscatter (e.g., Peruvian Amazon case study).

Given the particular vulnerability of Rhodes Islands landscape and the increasing frequency and severity of fires in Mediterranean regions under climate change, our study aims to explore and analyse Sentinel-1 SAR imagery to characterise the spatial and temporal footprint of the 2023 wildfire. Specifically, we examine pre- and post-fire Sentinel-1 scenes, compute backscatter change metrics and map the derived burnscar extents and potentially altered surface conditions. The scientific relevance of this work lies in improving our understanding of wildfire impacts in Mediterranean insular systems, and the environmental relevance derives from providing timely information that can support post-fire land-management, hazard mitigation (such as erosion control) and resilience planning.

Four bibliographic references underpin this approach: one on near-real-time wildfire monitoring using Sentinel-1 backscatter time series (Ban et al., 2020), one on continuous fire-induced vegetation loss detection using Sentinel-1 SAR time series and Bayesian changepoint detection (Chvez et al., 2023), and one on the broader capacity of Sentinel-1 for environmental-change detection (Doblas et al., 2020).

# Materials and Methods

## I. Remote Sensing Data Acquisition

This study utilized satellite imagery from both the Sentinel-1 and Sentinel-2 missions, operated by the European Space Agency (ESA), to assess the impact of the 2023 wildfires on Rhodes Island, Greece. The geographical area of interest was defined by a bounding box encompassing Rhodes Island (coordinates: [[27.6, 35.8], [27.6, 36.5], [28.4, 36.5], [28.4, 35.8], [27.6, 35.8]]). Two distinct temporal periods were established for 'before' and 'after' fire analysis: a pre-fire period from July 1, 2023, to July 15, 2023, and a post-fire period from July 28, 2023, to August 15, 2023.

**Sentinel-2 Optical Data:** Sentinel-2 Level-2A (Bottom-Of-Atmosphere reflectance) imagery, acquired from the 'COPERNICUS/S2_SR_HARMONIZED' collection, was used for optical analysis. Key bands utilized include B8 (Near-Infrared, NIR), B4 (Red), B3 (Green), B2 (Blue), and B12 (Short-Wave Infrared 2, SWIR2). These bands offer spatial resolutions of 10m for NIR, Red, Green, and Blue bands, and 20m for SWIR2. All Sentinel-2 images were filtered to include only those with a cloud pixel percentage of less than 20% over the area of interest.

**Sentinel-1 Synthetic Aperture Radar (SAR) Data:** Sentinel-1 Ground Range Detected (GRD) products were sourced from the 'COPERNICUS/S1_GRD' collection. The analysis focused on Interferometric Wide (IW) swath mode data, specifically on descending orbit passes, which ensured consistent viewing geometry. Dual polarization bands, Vertical-Vertical (VV) and Vertical-Horizontal (VH), were selected due to their sensitivity to surface roughness and vegetation structure, respectively. Sentinel-1 GRD products have a spatial resolution of approximately 10 meters.

## II. Tools and Libraries

All data processing and analysis were conducted within a Google Colaboratory environment, leveraging Python as the primary programming language. The core tools and libraries employed were:

*   **Google Earth Engine (GEE):** The `ee` Python client library facilitated programmatic access to GEE's extensive catalog of satellite imagery and its powerful cloud-based computational capabilities for large-scale geospatial analysis.
*   **Geemap:** The `geemap` Python package, built upon GEE and `ipyleaflet` or `folium`, was used for interactive visualization of geospatial data within the Jupyter notebook environment, allowing for dynamic mapping and exploration of results.

## III. Principles of Applied Methods

To characterize the wildfire impact, the following methods were applied to both Sentinel-1 and Sentinel-2 data:

1.  **Temporal Compositing:** For both Sentinel-1 and Sentinel-2, a median composite was generated for the 'before' and 'after' periods. This approach helps to minimize atmospheric effects (for Sentinel-2) and speckle noise (for Sentinel-1) by taking the median pixel value over the defined time window.

2.  **Cloud Masking (Sentinel-2):** A custom cloud masking function was applied to Sentinel-2 imagery. This function utilized the QA60 band, which provides bitmasks for cloud and cirrus presence, to effectively remove cloudy pixels and ensure clearer surface observations.

3.  **Speckle Filtering (Sentinel-1):** To mitigate the inherent speckle noise in SAR imagery, a `focal_median` filter with a 3x3 kernel was applied to Sentinel-1 images. This basic filter helps to smooth the data, making features more discernible without excessively blurring edges.

4.  **Optical Change Detection (Normalized Burn Ratio - NBR and Differenced NBR - dNBR):**
    *   The Normalized Burn Ratio (NBR) was calculated for both pre- and post-fire Sentinel-2 composites using the formula: NBR = (NIR - SWIR2) / (NIR + SWIR2), where NIR corresponds to B8 and SWIR2 to B12. NBR is a spectral index sensitive to the amount of green vegetation and water content.
    *   The Differenced Normalized Burn Ratio (dNBR) was then computed by subtracting the post-fire NBR from the pre-fire NBR (dNBR = NBR_before - NBR_after). Positive dNBR values indicate vegetation loss or stress, with higher values correlating to increased burn severity. This allows for quantitative assessment and mapping of burn severity categories (e.g., unburned, low, moderate, high severity).

5.  **SAR Change Detection (Logarithmic Ratio and Polarimetric Ratio Change):**
    *   **VH Logarithmic Ratio:** Change detection using Sentinel-1 data focused on the VH polarization band. A logarithmic ratio of the post-fire VH backscatter to the pre-fire VH backscatter was calculated: `10 * log10(VH_after / VH_before)`. A decrease in VH backscatter (negative dB values) is a strong indicator of changes in vegetation structure, consistent with fire damage.
    *   **VV/VH Polarimetric Ratio Change:** The VV/VH ratio was calculated for both pre- and post-fire images, and the difference (`Ratio_after - Ratio_before`) was computed. Changes in this ratio are sensitive to structural alterations in vegetation, providing an additional dimension for identifying modified zones.

6.  **Affected Area Extraction (Sentinel-1):** Burned areas were quantitatively estimated by applying a threshold to the Sentinel-1 VH change (dB) image. Pixels with a change below a defined threshold (e.g., -0.5 dB) were classified as burned, forming a binary mask. Burn severity categories (e.g., unburned, low, moderate, high severity) were defined based on thresholds commonly established in scientific literature, often influenced by USGS guidelines, allowing for quantitative assessment and mapping The total area of these masked pixels within the Area of Interest was then calculated in square kilometers using GEE's `pixelArea` function.

Specific functions for cloud masking (for Sentinel-2), NBR calculation, SAR speckle filtering, and various change detection methods were developed and adapted within the notebook.

```python
import ee
import geemap # Uncomment if not already imported and initialized

# --- Define Region of Interest (Rhodes Island, Greece) ---
# A central point for Rhodes Island, Greece (example coordinates)
aoi_center_coords = [28.0, 36.15]

# A bounding box for Rhodes Island (adjust as needed for specific wildfire extent)
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

# --- Define Dates for Before and After Wildfire (Rhodes, Summer 2023) ---
# Adjust these dates based on specific wildfire events on Rhodes in Summer 2023
before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-2 Cloud Masking Function ---
def mask_s2_clouds(image):
    qa = image.select('QA60')

    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11

    # Both flags should be set to zero, indicating clear conditions.
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0).And(
           qa.bitwiseAnd(cirrus_bit_mask).eq(0))

    return image.updateMask(mask).divide(10000)

# --- Load Sentinel-2 Surface Reflectance (SR) Collection ---
def get_s2_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))

    # Apply cloud mask
    masked_collection = collection.map(mask_s2_clouds)

    # Select bands for False Color Infrared (B8:NIR, B4:Red, B3:Green)
    # and compute median composite
    composite = masked_collection.median().select('B8', 'B4', 'B3')
    return composite

# Get 'before' and 'after' composites
image_before = get_s2_composite(before_start_date, before_end_date, aoi_polygon)
image_after = get_s2_composite(after_start_date, after_end_date, aoi_polygon)

# --- Visualization Parameters (False Color Infrared) ---
fci_vis_params = {
  'min': 0.0,
  'max': 0.3,
  'bands': ['B8', 'B4', 'B3']
}

# --- Display on an interactive map ---
# Check if geemap.Map() is already initialized or uncommented
# If you've commented out geemap in your notebook, uncomment it before running this.

Map = geemap.Map() # Uncomment this line if Map is not initialized
Map.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9) # Uncomment this line

# Add layers to the map
if 'Map' in locals() or 'Map' in globals(): # Check if Map object exists
    Map.addLayer(image_before.clip(aoi_polygon), fci_vis_params, 'Before Wildfire (FCI)')
    Map.addLayer(image_after.clip(aoi_polygon), fci_vis_params, 'After Wildfire (FCI)')
    Map.centerObject(aoi_polygon, 10) # Center map on the AOI at zoom level 10
    display(Map)
else:
    print("geemap.Map() is not initialized. Please ensure 'import geemap' and 'Map = geemap.Map()' are uncommented and run.")
    print("Alternatively, you can get the image URLs directly from Earth Engine:")
    # Example of how to get tile URLs if geemap is not used (for debugging/alternative display)
    # url_before = image_before.getMapId(fci_vis_params)
    # print(f"Before wildfire image URL: {url_before['tile_fetcher'].url_format}")
    # url_after = image_after.getMapId(fci_vis_params)
    # print(f"After wildfire image URL: {url_after['tile_fetcher'].url_format}")

print("Note: The dates used are placeholders for the current year's summer. Adjust these to the specific dates of the wildfire in Attique during summer 2024 once the event occurs and data is available.")
```

### Calculating and Visualizing Normalized Burn Ratio (NBR) and Differenced NBR (dNBR)

This section calculates the Normalized Burn Ratio (NBR) before and after the wildfire, and then computes the Differenced Normalized Burn Ratio (dNBR) to assess burn severity. dNBR is a common index used to quantify the ecological impact of wildfires using satellite imagery.


```python
import ee
import geemap # Ensure geemap is imported for map functionality

# Reuse the AOI and dates from the previous analysis for Rhodes
aoi_center_coords = [28.0, 36.15]
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-2 Cloud Masking Function ---
def mask_s2_clouds(image):
    qa = image.select('QA60')

    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11

    # Both flags should be set to zero, indicating clear conditions.
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0).And(
           qa.bitwiseAnd(cirrus_bit_mask).eq(0))

    return image.updateMask(mask).divide(10000)

# --- Load Sentinel-2 Surface Reflectance (SR) Collection ---
def get_s2_composite_for_nbr(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))

    # Apply cloud mask
    masked_collection = collection.map(mask_s2_clouds)

    # Select bands for NBR (NIR: B8, SWIR2: B12)
    # and compute median composite
    composite = masked_collection.median().select('B8', 'B12')
    return composite

# --- Calculate NBR Function ---
def calculate_nbr(image):
    # NBR = (NIR - SWIR2) / (NIR + SWIR2)
    return image.normalizedDifference(['B8', 'B12']).rename('NBR')

# Get 'before' and 'after' Sentinel-2 composites for NBR calculation
image_before_nbr_bands = get_s2_composite_for_nbr(before_start_date, before_end_date, aoi_polygon)
image_after_nbr_bands = get_s2_composite_for_nbr(after_start_date, after_end_date, aoi_polygon)

# Calculate NBR for before and after images
nbr_before = calculate_nbr(image_before_nbr_bands)
nbr_after = calculate_nbr(image_after_nbr_bands)

# Calculate Differenced Normalized Burn Ratio (dNBR)
dnbr = nbr_before.subtract(nbr_after).rename('dNBR')

# --- Visualization Parameters for dNBR ---
dnbr_vis_params = {
  'min': -0.5,
  'max': 0.7,
  'palette': [
      'green',   # Unburned / vegetation recovery
      'yellow',  # Low severity
      'orange',  # Moderate severity
      'red',     # High severity
      'maroon'   # Very high severity
  ]
}

# --- Display on a new interactive map ---
Map_dNBR = geemap.Map()
Map_dNBR.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9)

# Add the dNBR layer to the map
Map_dNBR.addLayer(dnbr.clip(aoi_polygon), dnbr_vis_params, 'dNBR Burn Severity')
Map_dNBR.centerObject(aoi_polygon, 10)

display(Map_dNBR)
```

### Visualizing Sentinel-1 SAR Data in Grayscale (Classic SAR Image Style)

This section **visualizes** the Sentinel-1 SAR data for the Rhodes wildfires, focusing on a single polarization band (VV) rendered in grayscale to provide a 'classic SAR image' appearance. This allows for a different perspective on the radar signal's interaction with the landscape.


```python
import ee
import geemap # Ensure geemap is imported for map functionality

# Reuse the AOI and dates from the previous analysis for Rhodes
aoi_center_coords = [28.0, 36.15]
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-1 Data Processing Function (re-defining for self-contained cell) ---
def get_s1_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.eq('instrumentMode', 'IW')) \
                   .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VH'))

    # Select VV and VH polarization bands
    collection = collection.select(['VV', 'VH'])

    # Apply speckle filter (focal median for simplicity)
    # A more advanced filter like RefinedLee could be used for better results
    def speckle_filter(image):
        # Apply a 3x3 median filter to reduce speckle noise
        return image.focal_median(3, 'square', 'pixels')

    filtered_collection = collection.map(speckle_filter)

    # Compute median composite
    composite = filtered_collection.median()
    return composite

# Get 'before' and 'after' Sentinel-1 composites
image_before_s1 = get_s1_composite(before_start_date, before_end_date, aoi_polygon)
image_after_s1 = get_s1_composite(after_start_date, after_end_date, aoi_polygon)

# --- Visualization Parameters for Grayscale SAR (VV band) ---
sar_grayscale_vis_params = {
  'min': -20, # Typical min dB value for SAR
  'max': 0,   # Typical max dB value for SAR
  'bands': ['VV'], # Display only the VV band
  'palette': ['black', 'white'] # Grayscale palette
}

# --- Display on a new interactive map ---
Map_s1_grayscale = geemap.Map()
Map_s1_grayscale.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9)

# Add layers to the map
Map_s1_grayscale.addLayer(image_before_s1.clip(aoi_polygon), sar_grayscale_vis_params, 'Before Wildfire (S1 VV - Grayscale)')
Map_s1_grayscale.addLayer(image_after_s1.clip(aoi_polygon), sar_grayscale_vis_params, 'After Wildfire (S1 VV - Grayscale)')
Map_s1_grayscale.centerObject(aoi_polygon, 10)

display(Map_s1_grayscale)
```

### Exploring Wildfire Impact with Sentinel-1 SAR Polarimetric Ratio Change

This section calculates and visualizes the change in the VV/VH polarimetric ratio from Sentinel-1 data. This ratio can be sensitive to structural changes in vegetation, providing another dimension for identifying and understanding wildfire-affected areas.

```python
import ee
import geemap # Ensure geemap is imported for map functionality

# Reuse the AOI and dates from the previous analysis for Rhodes
aoi_center_coords = [28.0, 36.15]
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-1 Data Processing Function (re-defining for self-contained cell) ---
def get_s1_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.eq('instrumentMode', 'IW')) \
                   .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VH'))

    # Select VV and VH polarization bands
    collection = collection.select(['VV', 'VH'])

    # Apply speckle filter (focal median for simplicity)
    def speckle_filter(image):
        # Apply a 3x3 median filter to reduce speckle noise
        return image.focal_median(3, 'square', 'pixels')

    filtered_collection = collection.map(speckle_filter)

    # Compute median composite
    composite = filtered_collection.median()
    return composite

# Get 'before' and 'after' Sentinel-1 composites
image_before_s1 = get_s1_composite(before_start_date, before_end_date, aoi_polygon)
image_after_s1 = get_s1_composite(after_start_date, after_end_date, aoi_polygon)

# --- Calculate VV/VH Ratio Change ---
vv_before = image_before_s1.select('VV')
vh_before = image_before_s1.select('VH')
vv_after = image_after_s1.select('VV')
vh_after = image_after_s1.select('VH')

# Calculate VV/VH ratio for before and after
ratio_before = vv_before.divide(vh_before).rename('VV_VH_Ratio_Before')
ratio_after = vv_after.divide(vh_after).rename('VV_VH_Ratio_After')

# Calculate the change in VV/VH ratio
ratio_change = ratio_after.subtract(ratio_before).rename('VV_VH_Ratio_Change')

# --- Visualization Parameters for VV/VH Ratio Change ---
# A diverging palette is good for changes (negative, no change, positive)
ratio_change_vis = {
  'min': -0.5, # Adjust based on typical ratio change values
  'max': 0.5,  # Adjust based on typical ratio change values
  'palette': ['blue', 'white', 'red'] # Blue for decrease, white for no change, red for increase
}

# --- Display on a new interactive map ---
Map_s1_ratio_change = geemap.Map()
Map_s1_ratio_change.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9)

# Add the VV/VH Ratio Change layer to the map
Map_s1_ratio_change.addLayer(ratio_change.clip(aoi_polygon), ratio_change_vis, 'SAR VV/VH Ratio Change')

# Optionally, add the 'after' composite for context
Map_s1_ratio_change.addLayer(image_after_s1.clip(aoi_polygon), {'min': -20, 'max': 0, 'bands': ['VV', 'VH'], 'gamma': 1.6}, 'After Wildfire (S1 VV,VH) for Context', False) # Set visible to False by default

Map_s1_ratio_change.centerObject(aoi_polygon, 10)

display(Map_s1_ratio_change)
```

### Visualizing Wildfire Impact with Sentinel-1 SAR Data (Rhodes, Summer 2023)

This section uses Sentinel-1 SAR data to detect changes caused by the Rhodes wildfires. We'll specifically look at the VH polarization, which is sensitive to vegetation structure, and calculate a change map.


```python
import ee
import geemap # Ensure geemap is imported for map functionality

# Reuse the AOI and dates from the previous analysis for Rhodes
aoi_center_coords = [28.0, 36.15]
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-1 Data Processing Function ---
def get_s1_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.eq('instrumentMode', 'IW')) \
                   .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VH'))

    # Select VV and VH polarization bands
    collection = collection.select(['VV', 'VH'])

    # Apply speckle filter (focal median for simplicity)
    # A more advanced filter like RefinedLee could be used for better results
    def speckle_filter(image):
        # Apply a 3x3 median filter to reduce speckle noise
        return image.focal_median(3, 'square', 'pixels')

    filtered_collection = collection.map(speckle_filter)

    # Compute median composite
    composite = filtered_collection.median()
    return composite

# Get 'before' and 'after' Sentinel-1 composites
image_before_s1 = get_s1_composite(before_start_date, before_end_date, aoi_polygon)
image_after_s1 = get_s1_composite(after_start_date, after_end_date, aoi_polygon)

# --- Calculate Change Detection ---
# Focus on VH band for burn scar detection; often shows decrease in backscatter
vh_before = image_before_s1.select('VH')
vh_after = image_after_s1.select('VH')

# Calculate log ratio (change in dB)
# A negative value indicates a decrease in backscatter (potential burn scar)
change_vh_db = vh_after.divide(vh_before).log10().multiply(10)

# --- Visualization Parameters for SAR ---
sar_vis_params = {
  'min': -20,
  'max': 0,
  'bands': ['VV', 'VH'],
  'gamma': 1.6
}

# Visualization for the change map (log ratio of VH)
change_vis_params = {
  'min': -5, # Significant decrease in backscatter
  'max': 5,  # Significant increase in backscatter
  'palette': ['blue', 'white', 'red'] # Blue for decrease, white for no change, red for increase
}

# --- Display on a new interactive map ---
Map_s1 = geemap.Map()
Map_s1.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9)

# Add layers to the map
Map_s1.addLayer(image_before_s1.clip(aoi_polygon), sar_vis_params, 'Before Wildfire (S1 VV,VH)')
Map_s1.addLayer(image_after_s1.clip(aoi_polygon), sar_vis_params, 'After Wildfire (S1 VV,VH)')
Map_s1.addLayer(change_vh_db.clip(aoi_polygon), change_vis_params, 'SAR VH Change (dB)')
Map_s1.centerObject(aoi_polygon, 10)

display(Map_s1)
```

### Quantitative Estimation of Burned Area from Sentinel-1 SAR Data

This section provides a quantitative estimate of the area burned during the Rhodes wildfire event, based on the Sentinel-1 VH change detection mask. The area is calculated in square kilometers.

```python
import ee
import geemap # Ensure geemap is imported for map functionality

# Reuse the AOI and dates from the previous analysis for Rhodes
aoi_center_coords = [28.0, 36.15]
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15' # Just before major wildfires started

after_start_date = '2023-07-28'  # After major wildfires
after_end_date = '2023-08-15'

# --- Sentinel-2 Cloud Masking Function (re-defining for self-contained cell) ---
def mask_s2_clouds(image):
    qa = image.select('QA60')

    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11

    # Both flags should be set to zero, indicating clear conditions.
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0).And(
           qa.bitwiseAnd(cirrus_bit_mask).eq(0))

    return image.updateMask(mask).divide(10000)

# --- Load Sentinel-2 Surface Reflectance (SR) Collection (re-defining for self-contained cell) ---
def get_s2_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))

    # Apply cloud mask
    masked_collection = collection.map(mask_s2_clouds)

    # Select bands for True Color (B4:Red, B3:Green, B2:Blue)
    # and compute median composite
    composite = masked_collection.median().select('B4', 'B3', 'B2')
    return composite

# Get 'before' and 'after' composites in True Color
image_before_tc = get_s2_composite(before_start_date, before_end_date, aoi_polygon)
image_after_tc = get_s2_composite(after_start_date, after_end_date, aoi_polygon)

# --- Visualization Parameters (True Color) ---
true_color_vis_params = {
  'min': 0.0,
  'max': 0.3,
  'bands': ['B4', 'B3', 'B2']
}

# --- Display on a new interactive map ---
Map_tc = geemap.Map()
Map_tc.setCenter(aoi_center_coords[0], aoi_center_coords[1], 9)

# Add layers to the map
Map_tc.addLayer(image_before_tc.clip(aoi_polygon), true_color_vis_params, 'Before Wildfire (True Color)')
Map_tc.addLayer(image_after_tc.clip(aoi_polygon), true_color_vis_params, 'After Wildfire (True Color)')
Map_tc.centerObject(aoi_polygon, 10)

display(Map_tc)
```

```python
import ee

# Reuse the AOI and dates from previous analysis
aoi_polygon = ee.Geometry.Polygon(
    [[[27.6, 35.8],
      [27.6, 36.5],
      [28.4, 36.5],
      [28.4, 35.8],
      [27.6, 35.8]]])

before_start_date = '2023-07-01'
before_end_date = '2023-07-15'

after_start_date = '2023-07-28'
after_end_date = '2023-08-15'

# --- Sentinel-1 Data Processing Function (copied for self-containment) ---
def get_s1_composite(start_date, end_date, aoi):
    collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
                   .filterDate(start_date, end_date) \
                   .filterBounds(aoi) \
                   .filter(ee.Filter.eq('instrumentMode', 'IW')) \
                   .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
                   .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VH'))

    collection = collection.select(['VV', 'VH'])

    def speckle_filter(image):
        return image.focal_median(3, 'square', 'pixels')

    filtered_collection = collection.map(speckle_filter)
    composite = filtered_collection.median()
    return composite

# Get 'before' and 'after' Sentinel-1 composites
image_before_s1 = get_s1_composite(before_start_date, before_end_date, aoi_polygon)
image_after_s1 = get_s1_composite(after_start_date, after_end_date, aoi_polygon)

# Calculate VH change (log ratio)
vh_before = image_before_s1.select('VH')
vh_after = image_after_s1.select('VH')
change_vh_db = vh_after.divide(vh_before).log10().multiply(10)

# Define the burn threshold (same as used in previous visualization for consistency)
burn_threshold = -0.5 # dB

# Create a binary mask where values below the threshold are considered burn scars
burn_mask = change_vh_db.lt(burn_threshold).selfMask()

# --- Calculate the area of the burn mask ---
# Get the area of each pixel in square meters
pixel_area = ee.Image.pixelArea()

# Multiply the burn mask by the pixel area to get area of burned pixels
burned_area_image = burn_mask.multiply(pixel_area)

# Sum the areas of all burned pixels within the AOI
total_burned_area_sqm = burned_area_image.reduceRegion(
    reducer=ee.Reducer.sum(),
    geometry=aoi_polygon,
    scale=10, # Sentinel-1 resolution is approx 10m
    maxPixels=1e9 # Increase maxPixels for larger areas if needed
).get('VH') # Get the sum from the 'VH' band, as that's what the mask is based on

# Convert to square kilometers and print
if total_burned_area_sqm.getInfo() is not None:
    total_burned_area_sqkm = ee.Number(total_burned_area_sqm).divide(1e6) # Convert sqm to sqkm
    print(f"Estimated Burned Area (Sentinel-1 VH change < {burn_threshold} dB): {total_burned_area_sqkm.getInfo():.2f} square kilometers")
else:
    print("Could not estimate burned area. Check if the burn mask is valid or if the AOI contains masked pixels.")
```

#Discussion

This study employed a multi-sensor approach, utilizing Sentinel-1 Synthetic Aperture Radar (SAR) and Sentinel-2 optical imagery, to comprehensively characterize the spatial extent and impact of the 2023 Rhodes Island wildfire. The integration of both data types provided a robust framework for assessing the event, each offering unique strengths for change detection and characterization.

## Interpretation and Comparison of Results

The Sentinel-2 optical data provided highly interpretable visual evidence of the wildfire's impact. False Color Infrared (FCI) composites clearly depicted the dramatic transformation of the landscape, with healthy, pre-fire vegetation (bright red) being replaced by stark, dark burn scars post-event. This visual contrast was further quantified through the Differenced Normalized Burn Ratio (dNBR), which effectively mapped burn severity across the island. High positive dNBR values, represented by orange, red, and maroon tones, directly corresponded to areas of significant vegetation loss and high burn severity, offering a partial but indicative assessment of ecological damage. Comparing the 'before' and 'after' FCI and dNBR maps unequivocally demonstrated the widespread destruction caused by the wildfire, particularly in forested and shrubland ecosystems.

Sentinel-1 SAR data offered a complementary perspective, focusing on physical and structural changes. The VH polarization backscatter, highly sensitive to vegetation volume, showed significant decreases (negative dB change) in burned areas, which were distinctly highlighted as blue regions in the SAR VH change detection maps. The false-color change composite (R:VH_after, G:VV_after, B:VH_before) visually reinforced this, presenting burn scars with a characteristic blue hue. Furthermore, changes in the VV/VH polarimetric ratio provided additional insights into modifications of surface roughness and vegetation architecture, which can be altered by fire. These SAR products allowed for the identification of modified zones, even under conditions that would render optical data unusable, providing crucial information on the immediate aftermath of the event.

## Critical Analysis of Data Capacity

Both Sentinel-1 and Sentinel-2 data proved instrumental, each with distinct advantages and limitations. Sentinel-2, with its visible and infrared bands (10m to 20m resolution), offers high sensitivity to vegetation health and chlorophyll content, making indices like NBR and dNBR powerful tools for directly assessing vegetation stress and burn severity. Its primary limitation, however, lies in its reliance on clear atmospheric conditions. During active fire events, smoke plumes, and subsequent cloud cover can significantly obscure the ground, delaying or preventing crucial post-fire assessments.

Sentinel-1 SAR, conversely, operates independently of sunlight and can penetrate clouds and smoke, making it exceptionally valuable for rapid response and monitoring during and immediately after a fire. Its 10m resolution provides detailed spatial information on structural changes. The sensitivity of VH backscatter to vegetation structure is a key advantage for burn scar detection. However, SAR data interpretation is often less intuitive than optical imagery, requiring specialized knowledge and processing (e.g., speckle filtering, terrain correction). The signal can also be affected by factors such as soil moisture, which may introduce complexity in differentiating burn scars from other land surface changes.

## Difficulties, Lessons Learned, and Contribution

A significant difficulty encountered was the sensitivity of quantitative burned area estimation to the chosen backscatter change threshold. Varying the threshold for Sentinel-1 VH change (e.g., from -2.5 dB to -0.5 dB) led to substantial differences in the calculated burned area (from 1.35 km to 126.39 km). This highlights that while SAR provides robust change detection, the interpretation and quantification of impact require careful consideration of thresholds, often necessitating validation with ground truth or complementary data. The lessons learned underscore the importance of a multi-sensor approach; Sentinel-2 provides detailed spectral insights into burn severity, while Sentinel-1 offers unparalleled temporal coverage and robustness against atmospheric interference.

The processing conducted here significantly contributes to the objective of characterizing the Rhodes Island wildfire by providing diverse, interpretable, and quantifiable measures of its impact. The generated maps and area estimates can support post-fire land management, hazard mitigation (e.g., erosion control planning), and resilience planning, offering timely information that is essential for effective environmental response and recovery efforts in Mediterranean insular systems susceptible to climate change-induced wildfire events.

This study significantly contributed to the characterization of the 2023 Rhodes Island wildfire by effectively employing a multi-sensor remote sensing approach, integrating both Sentinel-1 SAR and Sentinel-2 optical data. Essential points learned include the complementary nature of these data sources, where Sentinel-2 offers intuitive visual and quantitative spectral insights into burn severity (e.g., via dNBR), while Sentinel-1 provides robust, all-weather detection of structural changes (e.g., via VH change and polarimetric ratios). The analysis highlighted the critical influence of threshold selection on quantitative burned area estimations, underscoring the necessity for context-specific validation. Avenues for improvement and future perspectives involve exploring data fusion techniques to leverage the strengths of both optical and SAR data simultaneously, potentially through machine learning, and refining automated thresholding methods to enhance the accuracy and reliability of wildfire impact assessments in dynamic Mediterranean environments.
